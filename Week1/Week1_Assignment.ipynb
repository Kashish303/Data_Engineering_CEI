## Step 3 : Handle Missing Values

#### (i) Getting missing values

```python
print('Missing values in the dataset are')
print(df.isnull().sum())
```

```python
print('Only those Columns having missing values:')
missing = df.isnull().sum()
missing_df = pd.DataFrame({
    'Missing Count': missing
})
print(missing_df[missing_df['Missing Count']>0])

print('Total missing values are:')
print(df.isnull().sum().sum())
```

#### (ii) Handling missing values

```python
for col in [
    'discount',
    'what_customers_said',
    'seller_name',
    'videos',
    'seller_information',
    'variations'
]:
    print("\n", col)
    print(df[col].dtype)
    print(df[col].dropna().head())
```

#### Data Cleaning Strategy

Each column was handled based on its domain meaning,
not generic rules:

- discount → 0 (absence of discount is not average discount)
- JSON columns → '[]' (maintains structural consistency)  
- seller_name → 'Unknown Seller' + missing flag (avoids false attribution)
- seller_information → 'Not Available' (human readable)

Generic approaches like mode/median were intentionally 
avoided where domain logic suggested otherwise.

```python
# Discount
df['discount'] = df['discount'].fillna(0)

# Customer data
df['what_customers_said'] = df['what_customers_said'].fillna('[]')

# Seller name
df['seller_name'] = df['seller_name'].fillna('Unknown Seller')

# Videos
df['videos'] = df['videos'].fillna('[]')

# Seller_information
df['seller_information'] = df['seller_information'].fillna('Unknown Seller')

# variations
df['variations'] = df['variations'].fillna('[]')
```

```python
# After cleaning missing values 
print(df.isnull().sum())
print(f'Total missing values are: {df.isnull().sum().sum()}')
```

```python
for col in [
    'discount',
    'what_customers_said',
    'seller_name',
    'videos',
    'seller_information',
    'variations'
]:
    print(f"\n--- {col} ---")
    print(df[col].head())
```

## Step 5 : Remove Duplicates

```python
print('Number of duplicate rows')
print(df.duplicated().sum())

print("Product ID duplicates:", df['product_id'].duplicated().sum())
```

```python
# To remove duplicates
df.drop_duplicates(inplace=True)
# if there is same order multiple times
df = df.drop_duplicates(
    subset='product_id',
    keep='first'
)
```

## Step 6 : Create Derived Column

```python
df['final_price'].unique()
```

```python
# Creating col quantity
df['quantity'] = 1

# Final price(in float)
price = (
    df['final_price']
    .str.replace('"', '', regex=False)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.strip()
    .astype(float)
)

#total amount
df['total_amount'] = price * df['quantity'] 
print(df.head())

# Checking if there are any null values
print('Null values are:')
print(df.isnull().sum().sum())
```
